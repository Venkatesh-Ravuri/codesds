```python
import pandas as pd
```

```python
data=pd.read_csv("rainfall in india 1901-2015.csv")
```

```python
data.groupby(['SUBDIVISION']).count()
```

```python
data.isna().sum()
```

```python
data1=data.drop(['ANNUAL','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'],axis=1)
```

```python
data1
```

```python
data2=data1.loc[(data.SUBDIVISION=='TELANGANA')]
```

```python
data2
```

```python
data3=data1.loc[(data.SUBDIVISION=='TELANGANA')]
```

```python
data3
```

```python
data3['ANNUAL RAIN']=data3.apply(lambda row:row.JAN+row.FEB+row.MAR+row.APR+row.MAY+row.JUN+row.JUL+row.AUG+row.SEP+row.OCT+row.NOV+row.DEC,axis=1)
```

```python
data3
```

```python
data4=data3.drop(['SUBDIVISION'],axis=1)
```

```python
data4
```

```python
cor=data4.corr()
cor
```

```python
!pip3 install seaborn
```

```python
import seaborn as sns
sns.heatmap(cor,vmax=1,vmin=-1,annot=True,linewidths=5,cmap='bwr')
```

```python
data4['SWM']=data4.apply(lambda row:row.JUN+row.JUL+row.AUG+row.SEP,axis=1)
```
